### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a função:
def f(x):
    return 2 * x + 3
```

```python
# Cria um conjunto de valores aleatórios de x (de 5 a 10):
x = np.linspace(5, 10, 6)
```

```python
# Calcula os valores de y para cada valor de x:
y = f(x)
```

```python
y
```

```python
# Cria o gráfico:
plt.figure(figsize = (6,4))
plt.plot(x,y, label = 'f(x) = 2x + 3')

# Adiciona um título e rótulos aos eixos:
plt.title('Gráfico da Função linear')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adiciona uma legenda:
plt.legend()

# Mostra o gráfico:
plt.grid(True)
plt.show
```

### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

```python
import math
def quadratica(a, b, c):
    delta = b**2 - 4 * a * c
    
    if delta > 0: 
        x1 = (-b + math.sqrt(delta))/ 2 * a
        x2 = (-b - math.sqrt(delta))/ 2 * a
        print(f"A equação possui 2 raízes reais: x1 = {x1} e x2 = {x2}.")
    elif delta < 0:
        print(f"A equação não possui raízes reais. As raízes pertences ao conjunto dos números complexos.")
    else:
        x = (-b + math.sqrt(delta))/ 2 * a
        print(f"A equação possui apenas uma raiz igual à {x}")
        
quadratica(1, -5, 6)
```

```python
# SOLUÇÃO PROPOSTA:
# PACOTE cmath - c de NÚMEROS COMPLEXOS - ELE UTIZA UM CONJUNTO DE DADOS MAIOR.
import cmath

a, b, c = 1, -5, 6

d = (b**2) - (4*a*c)
sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print('As raízes são {0} e {1}'.format(sol1, sol2))
```

### Exercício 7: Equação Quadrática com Entrada do Usuário

Modifique o programa anterior para aceitar coeficientes a, b, e c de um usuário e resolver a equação quadrática correspondente.

```python
def quadratica(a, b, c):
    delta = b**2 - 4 * a * c
 
    if delta > 0: 
        x1 = (-b + math.sqrt(delta))/ 2 * a
        x2 = (-b - math.sqrt(delta))/ 2 * a
        print(f"A equação possui 2 raízes reais e distintas: x1 = {x1} e x2 = {x2}.")
    elif delta < 0:
        print(f"A equação não possui raízes reais. As raízes pertences ao conjunto dos números complexos.")
    else:
        x = (-b + math.sqrt(delta))/ 2 * a
        print(f"A função tem uma única raiz real {x}.")
        
a = int(input('Digite o valor do coeficiente a: '))
b = int(input('Digite o valor do coeficiente b: '))
c = int(input('Digite o valor do coeficiente c: '))

quadratica(a, b, c)
```

### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

```python
def quadratica(a, b, c):
    delta = b**2 - 4 * a * c
 
    if delta > 0: 
        x1 = (-b + math.sqrt(delta))/ 2 * a
        x2 = (-b - math.sqrt(delta))/ 2 * a
        print(f"O discriminate (delta) é igual à {delta:.2f}.")
        print(f"A equação possui 2 raízes reais e distintas: x1 = {x1} e x2 = {x2}.")
    elif delta < 0:
        print(f"O discriminate (delta) é igual à {delta:.2f}.")
        print(f"A equação não possui raízes reais. As raízes pertences ao conjunto dos números complexos.")
    else:
        x = (-b + math.sqrt(delta))/ 2 * a
        print(f"O discriminate (delta) é igual à {delta:.2f}.")
        print(f"A função tem uma única raiz real igual à {x}.")
        
a = int(input('Digite o valor do coeficiente a: '))
b = int(input('Digite o valor do coeficiente b: '))
c = int(input('Digite o valor do coeficiente c: '))

quadratica(a, b, c)
```

### Exercício 9: Cálculo de Logaritmos

Escreva um script que calcula o logaritmo de base 10 de um número fornecido pelo usuário. Certifique-se de tratar o caso em que o logaritmo não pode ser definido (números negativos e zero).

```python
# Definindo a função logarítmica:
def funcao_log(x):
    if x <= 0:
        print('Para valores negativos ou zero, o cálculo não é possível no conjunto dos números reais.')
    else:
        resultado = np.log10(x)
        print(f"O logaritmo de base 10 do número {x} é igual à {resultado:.2f}.")
        
x = int(input('Digite um número para calcular o seu logarítmo de base 10: '))
funcao_log(x)
```

### Exercício 10: Aplicação de Logaritmos

Utilize logaritmo para calcular o tempo necessário para que um investimento atinja um determinado valor, considerando uma taxa de crescimento contínua.

'''
np.log10(x): Calcula o logaritmo de base 10.
np.log(x): Calcula o logaritmo natural (base e).
'''

```python
'''
Desenvolvendo a fórmula:
A = P.e^(rt) => A/P = e^(rt) => ln (A/P)=l ne^(rt) => ln e^x = x => ln (A/P) = rt => t = (ln (A/P))/r

''' 
def tempo_investimento(A, P, r):
    resultado = np.log(A/P) / (r /100)
    print(f"O tempo necessário para o valor investido no início de R${P:.2f} a uma taxa de crescimento contínuo {r:.2f}% atinja o valor de R${A:.2f} é de {resultado:.2f} anos.")

A = float(input('Digite o valor do investimento final em reais:'))
P = float(input('Digite o valor do investimento inicial em reais:'))
r = float(input('Digite o valor da taxa de crescimento ao ano em porcentagem:'))

tempo_investimento(A, P, r)
```
